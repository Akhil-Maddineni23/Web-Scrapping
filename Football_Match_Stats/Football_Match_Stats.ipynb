```python
import requests
from bs4 import BeautifulSoup #to parse the html page
import pandas as pd
import numpy as np
import json
```

```python
base_url = "https://understat.com/match/14836"
```

```python
base_url
```

```python
res = requests.get(base_url)
res
```

```python
soup = BeautifulSoup(res.content,'lxml')
scripts = soup.find_all('script')
```

```python
scripts
```

```python
#From all the above we need to get only the shots data..
scripts[1]
```

```python
shots_data = scripts[1].string
```

```python
shots_data
```

```python
type(shots_data)
```

```python
#Strip unncessary data from the start and the end such that we gona have only the json data
start_index = shots_data.index("('")+2
end_index = shots_data.index("')")
json_data = shots_data[start_index:end_index]
json_data = json_data.encode('utf8').decode('unicode_escape')
```

```python
#Convert string to json format
data = json.loads(json_data)
```

```python
data
```

```python
team1 = data['h']
team2 = data['a']
```
