```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://understat.com/match/14836"
```

```python
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content,'lxml')
soup
```

```python
players = soup.find_all('div',class_='player-name-container')
```

```python
for player in players:
    print(player.span.text)
```

```python
home_df = pd.DataFrame(columns=['Player Name','Player Position','Played Minutes'])
home = soup.find_all('div',class_='roster roster-home')
home_team = home[0].find('h3').a.text
labels = home[0].find_all('label',class_='player')
print(home_team)
for label in labels:
    player_name = label.div.text.strip()
    player_position = label.find('span',class_='player-position').text.strip()
    played_minutes = label.find('span',class_='player-time').text.strip()
    home_df = home_df.append({'Player Name':player_name,'Player Position':player_position,'Played Minutes':played_minutes},ignore_index=True)
```

```python
home_df
```

```python
away_df = pd.DataFrame(columns=['Player Name','Player Position','Played Minutes','Goals'])
away = soup.find_all('div',class_='roster roster-away')
away_team = away[0].find('h3').a.text
labels = away[0].find_all('label',class_='player')
print(away_team)
for label in labels:
    player_name = label.div.text.strip()
    player_position = label.find('span',class_='player-position').text.strip()
    played_minutes = label.find('span',class_='player-time').text.strip()
    away_df = away_df.append({'Player Name':player_name,'Player Position':player_position,'Played Minutes':played_minutes},ignore_index=True)
    
```
