```python
import pandas as pd
import numpy as np
import json
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.cricbuzz.com/cricket-team/india/2/players"
```

```python
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content,'lxml')
```

```python
players_div= soup.find('div',class_='cb-col-67 cb-col cb-left cb-top-zero')
squad_role = players_div.find_all('h3',class_='cb-sqd-rol')
squad_role = [x.text for x in squad_role]
squad_role
```

```python
Players = pd.DataFrame(columns =['Name','Stats_Link'])
```

```python
players = players_div.find_all('a',class_='cb-col cb-col-50')
```

```python
soup_parser = BeautifulSoup(html_content,'html.parser')
players_div= soup_parser.find('div',class_='cb-col-67 cb-col cb-left cb-top-zero')
players_tags = players_div.find_all('a',class_='cb-col cb-col-50')
```

```python
for i in range(len(players)):
    Players = Players.append({'Name':players[i].text,
                              'Stats_Link':players_tags[i]['href']},
                              ignore_index=True)
```

```python
Players_Personal_Info = pd.DataFrame(columns =['Name','Country','Born','Birth Place','Height','Role','Batting Style','Bowling Style','Career Info'])
```

```python
Players_Personal_Info
```

```python
main_link = 'https://www.cricbuzz.com'
```

```python
def Get_Personal_Info(link):
    url = main_link + link
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content,'lxml')
    
    data = soup.find_all('div',class_='cb-hm-rght')
    info = data[0].find_all('div',class_='cb-col cb-col-60 cb-lst-itm-sm')
    
    player = soup.find('div',class_='cb-col cb-col-80 cb-player-name-wrap')
    height = data[0].find('div',class_='cb-col cb-col-60').text
    arr =[]
    arr.append(player.h1.text)
    arr.append(player.h3.text)
    arr.append(info[0].text)
    arr.append(info[1].text)
    arr.append(height)
    arr.append(info[2].text)
    arr.append(info[3].text)
    if(len(info) == 6):
        arr.append(info[4].text)
        arr.append(info[5].text)
    else:
        arr.append("---")
        arr.append(info[4].text)
     
    return arr
   
```

```python
for link in Players['Stats_Link']:
    info =Get_Personal_Info(link)
     
    Players_Personal_Info = Players_Personal_Info.append({'Name':info[0],
                                                          'Country':info[1],
                                                          'Born':info[2],
                                                          'Birth Place':info[3],
                                                          'Height':info[4],
                                                          'Role':info[5],
                                                          'Batting Style': info[6],
                                                          'Bowling Style':info[7],
                                                          'Career Info': info[8]},
                                                           ignore_index=True)
    
```

```python
Players_Personal_Info
```

```python
bat_cols = ['Name','Format','M','In','NO','Runs','HS','Avg','BF','SR','100','200','50','4s','6s']
bowl_cols = ['Name','Format','M','In','B','Runs','Wkts','BBI','BBM','Econ','Avg','SR','5WH','10WH']
Team_Batting_Stats = pd.DataFrame(columns =bat_cols)
Team_Bowling_Stats = pd.DataFrame(columns =bowl_cols)
```

```python
Team_Batting_Stats
```

```python
Team_Bowling_Stats
```

```python
count =0
count1=0
for link in Players['Stats_Link']:
    url = main_link + link
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content,'lxml')
    
    player = soup.find('div',class_='cb-col cb-col-80 cb-player-name-wrap')
    name = player.h1.text
    name =[name]
    
    data = soup.find_all('div',class_='cb-hm-rght')
    tables = data[1].find_all('table',class_='table cb-col-100 cb-plyr-thead')
    
    #Batting Stats
    
    table_rows = tables[0].find_all('tr')
    for i in range(1,len(table_rows)):
        row = table_rows[i].find_all('td')
        row_values = [x.text for x in row]
        row_values = name + row_values
        Team_Batting_Stats.loc[count] = row_values
        count+=1
        
        
       
        
        
    #Bowling Stats
    table_rows1 = tables[1].find_all('tr')
    for i in range(1,len(table_rows1)):
        row1 = table_rows1[i].find_all('td')
        row_values1 = [x.text for x in row1]
        row_values1 = name + row_values1
        Team_Bowling_Stats.loc[count1] = row_values1
        count1+=1
```
